```python
df_final = df.drop(['Rank',	'Capital',	'Continent'], axis=1)
```

# **Population of India Over the Years**

```python
chosen_country = 'India'

country_data = df_final[df_final['Country/Territory'] == chosen_country]

years = country_data.columns[2:].str.extract('(\d+)')[0].astype(float).dropna().astype(int)
population = pd.Series(country_data.iloc[:, 2:].values.flatten())
valid_data = pd.DataFrame({'Year': years, 'Population': population}).dropna()
years = valid_data['Year'].astype(int).values

population = valid_data['Population'].astype(float).values
plt.figure(figsize=(10, 5))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.xticks(years, rotation=45)
plt.show()
```

```python
plt.hist(population, bins=30, color='brown', edgecolor='black')
plt.show()
```
